```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z= x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('My x label')
ax.set_ylabel('My y label')
ax.set_title('My x vs y graph')
```

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax2.set_title('Mysecond axis')
ax2.plot(x,y)
ax1.plot(x,z)
```

```python
fig
```

```python
fig3 = plt.figure()
ax1 = fig3.add_axes([0,0,1,1])
ax1.plot(x,z)
ax1.set_title('100% Zoom')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

ax2 = fig3.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,2))
```

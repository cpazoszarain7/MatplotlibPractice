```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe()
```

```python
tips.info()
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=30)
```

```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'], kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
plt.figure(figsize=(10,6))
sns.distplot(tips[tips['sex']=='Male']['total_bill'], bins=30, color='red', label="Male")
sns.distplot(tips[tips['sex']=='Female']['total_bill'],bins=30, label ='Female', rug=True)
plt.legend()
plt.show()
```

```python
tips.head()
```

```python
sns.pairplot(data=tips, y_vars=['tip'],x_vars=['total_bill','size', 'smoker'], hue='sex')
```

```python
sns.set_palette("YlGn")
sns.barplot(x=tips['sex'], y=tips['total_bill'])
```

```python
sns.countplot(tips['sex'])
```

```python
sns.boxplot(data=tips,x='day',y='total_bill', hue='sex')
```

```python
plt.figure(figsize=(12,6))
sns.distplot(tips[tips['day']=='Thur']['total_bill'], label="Thur")
sns.distplot(tips[tips['day']=='Fri']['total_bill'], label="Fri")
sns.distplot(tips[tips['day']=='Sat']['total_bill'], label="Sat")
sns.distplot(tips[tips['day']=='Sun']['total_bill'], label="Sun")
plt.legend()
plt.show()
```

```python
sns.violinplot(data=tips,x='day',y='tip', hue='sex', split=True)
```
